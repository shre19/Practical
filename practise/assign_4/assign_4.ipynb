###  medv has high positive corelation with rm and high negative corelation with lstat
### we choose lstat, indus, ptratio, rm, dis

```python
cols = ['lstat', 'indus', 'ptratio', 'rm', 'zn', 'dis', 'medv']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[cols])
df[cols] = scaler.transform(df[cols])
```

```python
df[cols].plot(kind='box', subplots=True, layout=(3,3), fontsize=18,figsize=(22,22))
```

```python
df.shape
```

```python
for i in cols:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    
    Lower_limit=Q1-1.5*IQR
    Upper_limit=Q3+1.5*IQR
    
    med = df.loc[(df[i]>lower_limit)&(df[i]<upper_limit), i].median()
    df[i] = np.where((df[i]<lower_limit)|(df[i]>upper_limit), med, df[i])
```

```python
df.shape
```

```python
df[cols].plot(kind='box', figsize=(22, 22), subplots=True, fontsize=18, layout=(3,3))
```

```python
features = df[cols[0:6]]
prices = df['medv']
```

```python
features
```

```python
prices
```

## Splitting the training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, prices, random_state=0, test_size=0.3)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Building the linear regression model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
for i in y_pred:
    print(i)
```

```python
y_test
```

```python
print("Training accuracy: ", reg.score(x_train, y_train))
```

```python
print("Testing accuracy: ", reg.score(x_test, y_test))
```
